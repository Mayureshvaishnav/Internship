```python
import pandas as pd
import numpy as np
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get ('https://en.wikipedia.org/wiki/Main_Page')
url
```

```python
soup = BeautifulSoup(url.content)
soup
```

```python
titles = soup.find_all(['h1', 'h2','h3','h4','h5','h6'])
print('List all the header tags :', *titles, sep='\n\n')
```
