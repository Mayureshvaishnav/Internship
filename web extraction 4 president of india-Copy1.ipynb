```python
import pandas as pd
import numpy as np
import bs4
from bs4 import BeautifulSoup
import requests
```

```python
url = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
url
```

```python
soup = BeautifulSoup(url.content)
soup
```

```python
A= []
for i in soup.find_all('div',class_ = 'presidentListing'):
    A.append(i.get_text().strip("\n").split("(")[0])
    
A
```

```python
d= []
for i in soup.find_all('div',class_ = 'presidentListing'):
    d.append(i.get_text())
    
d
```

```python
B= []
for i in soup.find_all('div',class_ = 'presidentListing'):
    B.append(i.get_text().split("\n")[2].strip("\n"))
    
B
```

```python
c= []
for i in soup.find_all('div',class_ = 'presidentListing'):
    c.append(i.get_text().split("\n")[2].strip("\n").replace('Term of Office:',''))
    
c
```

```python
df=  pd.DataFrame({'Former President of India': A, ' Term of Office ': c})
df
```
